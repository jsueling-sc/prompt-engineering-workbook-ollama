### Model choice

For this workbook we'll use [`granite4.1:3b`](https://ollama.com/ibm/granite4.1:3b),
a small open model from IBM. It's a ~2GB download, and to run it your machine
needs a few GB of free **memory** (RAM, or video memory if you have a GPU) on
top of the disk space for the download — the model's weights have to be held in
memory while it generates each response.

- The `3b` in the name is the rough size of the model: it has about 3 billion
  internal "parameters" (the numbers it learned during training). As a rule of
  thumb, all else equal, more parameters tends to mean a "smarter" model — but
  also a bigger, slower, more memory-hungry one.
- Even with the exact same prompt, **which model you pick has a large effect on
  the output**. A small model like this one will make mistakes a bigger model
  wouldn't — that's expected, and some exercises here are designed to show it.
- That said, the prompt-engineering ideas in this workbook apply to *any* large
  language model, and good prompting meaningfully improves quality, consistency,
  and reliability whatever model you run.
- **Optional:** Browse the [Ollama model catalogue](https://ollama.com/search)
  to see other open models you could try.

```python
import ollama

# Stay with `granite4.1:3b` for consistent results in this workbook
MODEL_NAME = "granite4.1:3b"

# Make sure the model is available locally (downloads it the first time),
# streaming progress so you can see what's happening on a first-time pull.
for progress in ollama.pull(MODEL_NAME, stream=True):
    status = progress.status or ""
    total = progress.total or 0
    completed = progress.completed or 0
    if total:
        pct = completed / total * 100
        print(f"\r{status}: {pct:5.1f}%  ({completed/1e9:.2f}/{total/1e9:.2f} GB)", end="", flush=True)
    else:
        print(f"\r{status}", end="", flush=True)
print()  # newline once the loop finishes

# Stores the MODEL_NAME variable for use across notebooks via the IPython store
%store MODEL_NAME

MODEL_NAME
```

Below is an example of what running a prompt will look like in this tutorial. First, we create `get_completion`, which is a helper function that sends a prompt to the Ollama server and returns a generated response from the model specified. Run that cell now.

```python
import ollama

def get_completion(prompt: str):
    response = ollama.chat(
        model=MODEL_NAME,
        messages=[{"role": "user", "content": prompt}],
        options={"temperature": 0.0, "num_predict": 2000},
    )
    return response["message"]["content"]
```

Now we will write out an example prompt and print the output by running our `get_completion` helper function. Running the cell below will print out a response from the model beneath it.

Feel free to play around with the prompt string to elicit different responses.

```python
# Prompt
prompt = "Hello World!"

# Get and print the model's response
print(get_completion(prompt))
```
