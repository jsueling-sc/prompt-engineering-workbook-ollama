## Setup

Run the cell below to load `MODEL_NAME` and define the `get_completion` helper.

````python
# Python's built-in regular expression library (used by the graders)
import re
import ollama

# Retrieve MODEL_NAME from the IPython store (set in Chapter 0).
%store -r MODEL_NAME

# num_predict is bumped to 4000 for this chapter so the long-output exercise has room.
def get_completion(user_prompt: str, system_prompt=""):
    messages = []
    if system_prompt:
        messages.append({"role": "system", "content": system_prompt})
    messages.append({"role": "user", "content": user_prompt})
    response = ollama.chat(
        model=MODEL_NAME,
        messages=messages,
        options={"temperature": 0.0, "num_predict": 4000},
    )
    return response["message"]["content"]
````

## Lesson

**Models respond best to clear, direct, and specific instructions.**

Treat the model like a brand-new colleague who has no context beyond the literal
words you give it. It can't read your mind, infer your unstated preferences, or
know which of several reasonable interpretations you meant. The more precisely
you spell out exactly what you want — the format, the length, the perspective,
what to leave out — the closer the response will land.

A useful test is the **Golden Rule of Clear Prompting**: hand your prompt to
another person and have them follow it literally. If *they* would be confused
about what you're asking for, the model will be too.

### Examples

#### Cutting the preamble

Models often wrap a task in a chatty introduction. Here we ask for a Python function and get a "Here's how you can implement this..." lead-in we may not want. Suppose that we want to directly execute whatever is returned in a Python [REPL](https://en.wikipedia.org/wiki/Read%E2%80%93eval%E2%80%93print_loop)

````python
USER_PROMPT = "Write a Python function named `reverse_string` that reverses a string."
print(get_completion(USER_PROMPT))
````

That response is helpful, but we can't run it. It comes wrapped in explanatory prose and a Markdown code fence` ```python `, so handing it straight to `exec()` would raise a `SyntaxError`.

If you just want the code itself, **ask for that explicitly.**

````python
USER_PROMPT = "Write a Python function named `reverse_string` that reverses a string. Skip any preamble or explanation and respond with only the function."

python_code = get_completion(USER_PROMPT)

print(python_code)
````

Now the response should be nothing but the function definition. We can pass that string directly to Python's built-in [`exec()`](https://docs.python.org/3/library/functions.html#exec), which runs it as code and defines the `reverse_string` function in our current environment.

````python
exec(python_code)
````

With the function now defined, we can call it like any other:

````python
reverse_string("Hello world!")
````

This round-trip — generate code, then run it untouched — only works because we stripped the preamble. The chatty version would have made `exec()` throw an error because of the prose. Being clear and direct turned the model's output into something directly usable.

> **If `exec()` still errors:** some models insist on wrapping code in Markdown fences even when told not to. Either add "Do not use Markdown formatting or code fences." to the prompt, or strip them first, e.g. `python_code = python_code.strip().removeprefix("```python").removesuffix("```")`.

#### Forcing a decision

Now let's ask an opinion question. Asked plainly, the model tends to hedge and
refuse to commit to a single answer.

````python
USER_PROMPT = "What is the best programming language?"
print(get_completion(USER_PROMPT))
````

We can get a committed answer simply by **removing its escape hatch** and
telling it to choose.

````python
USER_PROMPT = "What is the best programming language? I know it depends, but if you had to recommend exactly one language for a complete beginner to learn first, name that single language in your first sentence."
print(get_completion(USER_PROMPT))
````

### Exercise 2.1 - Shouting

Modify the `SYSTEM_PROMPT` so the model's entire answer comes back in **ALL
CAPITAL LETTERS**. The grader passes if every letter in the response is
uppercase.

````python
# System prompt - this is the only field you should change
SYSTEM_PROMPT = "[REPLACE THIS TEXT]"

# User prompt
USER_PROMPT = "What should I make for dinner tonight?"

response = get_completion(USER_PROMPT, SYSTEM_PROMPT)

# Function to grade exercise correctness
def grade_exercise(text):
    letters = [ch for ch in text if ch.isalpha()]
    return len(letters) > 0 and all(ch.isupper() for ch in letters)

print(response)
print("\n--------------------------- GRADING ---------------------------")
print("This exercise has been correctly solved:", "✅" if grade_exercise(response) else "❌")
````

### Exercise 2.2 - One Word Only

Modify `SYSTEM_PROMPT` so the model replies with **only the single word** that
answers it — no preamble, no explanation, no punctuation.

````python
# System prompt - this is the only field you should change
SYSTEM_PROMPT = "[REPLACE THIS TEXT]"

# User prompt
USER_PROMPT = "What's the capital of Japan"

response = get_completion(USER_PROMPT, SYSTEM_PROMPT)

# Function to grade exercise correctness
def grade_exercise(text):
    return text.strip() == "Tokyo"

print(response)
print("\n--------------------------- GRADING ---------------------------")
print("This exercise has been correctly solved:", "✅" if grade_exercise(response) else "❌")
````

### Exercise 2.3 - A Long Read

Modify the `PROMPT` so the model produces as long a response as you can get out
of it. If the answer is **800 words or more**, it grades as correct. (Models
aren't great at counting their own words, so aim well past the target.)

````python
# User prompt - this is the only field you should change
USER_PROMPT = "[REPLACE THIS TEXT]"

response = get_completion(USER_PROMPT)

# Function to grade exercise correctness
def grade_exercise(text):
    return len(text.strip().split()) >= 800

print(response)
print("\n--------------------------- GRADING ---------------------------")
print("This exercise has been correctly solved:", "✅" if grade_exercise(response) else "❌")
````
